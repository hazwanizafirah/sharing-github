```python
import pandas as pd
 
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
df = pd.read_csv('gs://normal-gas-turbine/Scatter.csv')
df= df.dropna()
df.head()
```

```python
x = range(100)
y = range(100,200)
fig = plt.figure()
ax1 = fig.add_subplot(111)


ax1.scatter(x=df['T5 Avg Turbine Temp_Normal'], y=df['Vib GB Accel Overall_Normal'], s=10, c='b', marker="s", label='healthy')
ax1.scatter(x=df['T5 Avg Turbine Temp_Problem'],y=df['Vib GB Accel Overall_Problem'], s=10, c='r', marker="o", label='unhealthy')
plt.legend(loc='upper left');
plt.show()
```
